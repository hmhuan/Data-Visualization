```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('./coronadata/18-5.csv')
```

```python
data.tail()
```

```python
data_by_country = data[data['#'].isna() == False]
```

## Top 10 Total cases by Country On 05-18-2020

```python
Ten_Biggest = data_by_country[:10]
```

### matplotlib

```python
plt.figure(figsize=(12, 5))
x = [(Ten_Biggest['Country, Other'][i] 
      + '\n(' + str(Ten_Biggest['Total Cases'][i]) 
      + ')') for i in Ten_Biggest.index]
y = Ten_Biggest['Total Cases']
plt.bar(x, y)

plt.xlabel("Country")
plt.ylabel("Total cases")
plt.title("Top 10 Total cases by Country\n(05-18-2020)")

plt.savefig('./output/Top_10_TTCases_by_Country.png')
plt.show()
```

### seaborn

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Country, Other', y='Total Cases', data=Ten_Biggest)
plt.savefig('./output/Top10bySNS.png')
```

```python
N = 10
plt.figure(figsize=(12, 5))
x1 = Ten_Biggest['Total Cases']
x2 = Ten_Biggest['Total Recovered']

ind = np.arange(N) 
width = 0.35       
plt.bar(ind, x1, width, label='Men')
plt.bar(ind + width, x2, width,
    label='Women')

plt.ylabel('Scores')
plt.title('Scores by group and gender')

plt.xticks(ind + width / 2, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend(loc='best')
plt.show()
```
