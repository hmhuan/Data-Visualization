```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('./coronadata/18-5.csv')
```

```python
data.tail()
```

```python
data_by_country = data[data['#'].isna() == False]
data_non_country = data[data['#'].isna()].reset_index()
```

## Top 10 Total cases by Country On 05-18-2020

```python
Ten_Biggest = data_by_country[:10]
```

### matplotlib

```python
plt.figure(figsize=(12, 7))
x = [(Ten_Biggest['Country, Other'][i] +'\n(' + str(Ten_Biggest['Total Cases'][i])+')') for i in Ten_Biggest.index]
y = Ten_Biggest['Total Cases']
plt.bar(x, y)
plt.xlabel("Country")
plt.ylabel("Total cases")
plt.title("Top 10 Total cases by Country\n(05-18-2020)")
# plt.savefig('./output/Top_10_TTCases_by_Country_plt.png')
plt.show()
```

### seaborn

```python
plt.figure(figsize=(12,7))
x = [(Ten_Biggest['Country, Other'][i] 
      + '\n(' + str(Ten_Biggest['Total Cases'][i]) 
      + ')') for i in Ten_Biggest.index]
y = Ten_Biggest['Total Cases']
g = sns.barplot(x='Country, Other', y='Total Cases', data={'Country, Other': x, 'Total Cases': y})
plt.xlabel("Country")
plt.ylabel("Total cases")
plt.title("Top 10 Total cases by Region\n(05-18-2020)")
# plt.savefig('./output/Top_10_TTCase_by_Country_SNS.png')
plt.show()
```

## Total cases by 6 regions

```python
six_regions = data_non_country[1:7]
```

### matplotlib

```python
plt.figure(figsize=(10, 7))
x = [(six_regions['Country, Other'][i] +'\n('+ str(six_regions['Total Cases'][i]) +')') for i in six_regions.index]
y = six_regions['Total Cases']
plt.bar(x, y)
plt.xlabel("Region")
plt.ylabel("Total cases")
plt.title("Total cases 6 Regions\n(05-18-2020)")
plt.savefig('./output/TTCases_Regions_plt.png')
plt.show()
```

### seaborn

```python
plt.figure(figsize=(10, 7))
x_ = [(six_regions['Country, Other'][i] +'\n('+ str(six_regions['Total Cases'][i]) +')') for i in six_regions.index]
y_ = six_regions['Total Cases']
sns.barplot(x='Country, Other', y='Total Cases', data={'Country, Other': x_, 'Total Cases': y_})
plt.xlabel("Region")
plt.ylabel("Total cases")
plt.title("Total cases 6 Regions\n(05-18-2020)")
plt.savefig('./output/TTCases_Regions_SNS.png')
plt.show()
```
